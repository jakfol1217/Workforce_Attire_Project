Notebook containing some basic exploration of the spanish-labeled dataset.

```python
import glob
from PIL import Image
import json
import os
```

```python
os.getcwd()
```

```python
json_files = glob.glob(os.path.join(os.getcwd(), "data/*.json"))
```

```python
genres = {'none':0}
classes = {'none':0}
for file in json_files:
    with open(file) as f:
        json_file = json.load(f)
    for box in json_file['arr_boxes']:
        if 'genre' in box:
            if box['genre'] in genres:
                genres[box['genre']] += 1
            else:
                genres[box['genre']] = 1
        else:
            genres['none'] += 1
        if 'class' in box:
            if box['class'] in classes:
                classes[box['class']] += 1
            else:
                classes[box['class']] = 1
        else:
            classes['none'] += 1
```

```python
Image.open(os.path.join(os.getcwd(), "dataset",json_file['file_name']))
```

```python
import matplotlib.pyplot as plt
```

```python
classes = {k: v for k, v in sorted(classes.items(), key=lambda item: item[1])}
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.barh(list(classes.keys()), list(classes.values()))
ax.set_title("Distribution of classes", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.bar(genres.keys(), genres.values())
plt.title("Distribution of genres", fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
